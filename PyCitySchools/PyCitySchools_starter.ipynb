```python
import pandas as pd
```

```python
schools="Resources/schools_complete.csv"
students="Resources/students_complete.csv" 
```

```python
school_read = pd.read_csv(schools)

school_read.head()
```

```python
student_read = pd.read_csv(students)

student_read.head()
```

```python
total_schools = school_read['school_name'].count()

total_schools
```

```python
total_students = student_read['Student ID'].count()

total_students
```

```python
total_budget = school_read['budget'].sum()

total_budget
```

```python
avg_math = student_read ['math_score'].mean()

avg_math
```

```python
avg_read = student_read ['reading_score'].mean()

avg_read
```

```python
pass_math = student_read.loc[student_read['math_score'] >= 70, ['math_score']]

pass_math.head(5)
```

```python
pass_math_count = pass_math['math_score'].count()
pass_math_count
```

```python
per_pass_math = pass_math_count/total_students
per_pass_math
```

```python
pass_read = student_read.loc[student_read['reading_score'] >= 70, ['reading_score']]

pass_read.head(5)
```

```python
pass_read_count = pass_read['reading_score'].count()
pass_read_count
```

```python

per_pass_read = pass_read_count/total_students
per_pass_read
```

```python
overall_passing = (per_pass_read + per_pass_math)/2

overall_passing
```

```python
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget":[total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_read], 
    "% Passing Math":[per_pass_math],
    "% Passing Reading":[per_pass_read],
    "Overall Passing Rate":[overall_passing],
})


district_summary = district_summary[["Total Schools",
                                     "Total Students",
                                     "Total Budget",
                                     "Average Math Score",
                                     "Average Reading Score", 
                                     "% Passing Math",
                                     "% Passing Reading",
                                     "Overall Passing Rate"]]
district_summary
```

```python
students_math_passing = student_read.loc[student_read['math_score']>70]
students_math_passing
```

```python
all_students_math_passing = pd.DataFrame({"school_name": students_math_passing["school_name"],
                                         "Count_students_passing_math": 0})

students_pass_math_school = all_students_math_passing.groupby(['school_name']).count().reset_index()

students_pass_math_school
```

```python
students_read_passing = student_read.loc[student_read['reading_score']>70]
students_read_passing
```

```python
all_students_read_passing = pd.DataFrame({"school_name":students_read_passing["school_name"],
                                         "Count_students_passing_read":0})

students_pass_read_school = all_students_read_passing.groupby(['school_name']).count().reset_index()

students_pass_read_school
```

```python
school_group = student_read.groupby(['school_name'])
```

```python
new_school_df = pd.merge(school_read, df_of_averages, on = 'school_name') \
.merge(students_pass_math_school, on = 'school_name') \
.merge(students_pass_read_school, on = 'school_name') 

new_school_df
```

```python
percent_passing_math_school = (new_school_df['Count_students_passing_math'] / new_school_df['size']) *100
percent_passing_math_school
```

```python
percent_passing_read_school = (new_school_df['Count_students_passing_read'] / new_school_df['size']) *100
percent_passing_read_school
```

```python
overall_passing_school = ((percent_passing_math_school + percent_passing_read_school) / 2)
overall_passing_school
```

```python
schools_summary_df = pd.DataFrame({
    "School Name": new_school_df['school_name'],
    "School Type": new_school_df['type'],
    "Total Students": new_school_df['size'],
    "Total School Budget": new_school_df['budget'],
    "Per Student Budget": (new_school_df['budget']/new_school_df['size']),
    "Average Math Score": new_school_df['Average Math Score'],
    "Average Reading Score": new_school_df['Average Reading Score'],
    "% Passing Math": percent_passing_math_school,
    "% Passing Reading": percent_passing_read_school,
    "Overall Passing Rate": overall_passing_school
})


schools_summary_df = schools_summary_df[["School Name",
                                     "School Type",
                                     "Total Students",
                                     "Total School Budget",
                                     "Per Student Budget", 
                                     "Average Math Score",
                                     "Average Reading Score",
                                     "% Passing Math",
                                     "% Passing Reading",
                                     "Overall Passing Rate"]]
schools_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_spending = schools_summary_df.copy()

bins = [0, 585, 615, 645, 675]

view_groups = ['$0-585', '$586-615', '$616-645', '$646-675']

view_groups
```

```python
school_spending['Per Student Budget']=school_spending['Per Student Budget']\
.replace('[\$]', '', regex=True).astype('float')

school_spending['Spending Ranges Per Student'] = pd.cut(school_spending['Per Student Budget'],bins, labels=view_groups)

school_spending
```

```python
school_spending = school_spending[['Average Math Score',
                                  'Average Reading Score',
                                  '% Passing Math',
                                  '% Passing Reading',
                                  'Overall Passing Rate',
                                  'Spending Ranges Per Student']]

average_scores_by_spending_df = school_spending.groupby('Spending Ranges Per Student').mean()

average_scores_by_spending_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
scores_by_size = schools_summary_df.copy()

bins = [0, 1500, 3500, 5000]

view_groups = ['Small (<1500) Students', 'Medium (1500-350) Students', 'Large (3500-5000) Students']

scores_by_size['Total Students'] = pd.to_numeric(scores_by_size['Total Students'])

scores_by_size['School Size'] = pd.cut(scores_by_size['Total Students'],
                                                                  bins, labels=view_groups)
scores_by_size = scores_by_size[['Average Math Score', 
                                 'Average Reading Score',
                                 '% Passing Math', 
                                 '% Passing Reading', 
                                 'Overall Passing Rate',
                                 'School Size']]

scores_by_size_df = scores_by_size.groupby('School Size').mean()

scores_by_size_df
```
